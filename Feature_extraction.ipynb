```python
import tensorflow
from keras.applications.resnet50 import ResNet50,preprocess_input
from keras.layers import GlobalMaxPooling2D
```

```python
model = ResNet50(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
model.trainable = False
```

```python
model=tensorflow.keras.Sequential([
    model,
    GlobalMaxPooling2D()
])
model.summary()
```

```python
def extract_feature(img_path,model):
    img=cv2.imread(img_path)
    if not os.path.exists(img_path):
        print(f"Error: Image file does not exist at {img_path}")
        return None
    img=cv2.resize(img,(224,224))
    img=np.array(img)
    exp_img=np.expand_dims(img,axis=0)
    pre_img=preprocess_input(exp_img)
    result=model.predict(pre_img).flatten()
    normalized=result/norm(result)
    return normalized
    
```

```python
extract_feature("C:/Users/shami/Desktop/Visual_search/Datasets/1636.jpg",model)
```

```python
import os
from tqdm import tqdm
```

```python
filename=[]
feature_list=[]

for file in os.listdir('Datasets'):
    filename.append(os.path.join('Datasets',file))
```

```python
filename[0:10]
```

```python
for file in tqdm(filename):
    feature_list.append(extract_feature(file,model))
```

```python
import pickle

# Open the file in write-binary mode and dump feature_list into it
with open("featurevector.pkl", "wb") as f:
    pickle.dump(feature_list, f)

# Open the file in write-binary mode and dump filename into it
with open("filenames.pkl", "wb") as f:
    pickle.dump(filename, f)
```
